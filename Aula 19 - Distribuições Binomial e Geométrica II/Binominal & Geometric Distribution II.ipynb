# Distribuição de Poisson

### Uso da PMF

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import linspace
from scipy.stats import poisson
import matplotlib.gridspec as gridspec


# Calculando a pmf
lambdas = [1, 4, 10]
outcome_counts = linspace(0, 10, 11)
outcome_probs = list()

for i in range(len(lambdas)):
    outcome_probs.append(poisson.pmf(outcome_counts, lambdas[i]))
    
fig = plt.figure(figsize=(20,4))

gs1 = gridspec.GridSpec(1, 3)

axes = list()
for i in range(len(outcome_probs)):
    axes.append(fig.add_subplot(gs1[i]))
    axes[i].bar(outcome_counts, outcome_probs[i])
    axes[i].set_title("Lambda = " + str(lambdas[i]))
    
plt.show()
```

### Uso da CDF

```python
outcome_cdf = list()
for i in range(len(lambdas)):
    outcome_cdf.append(poisson.cdf(outcome_counts, lambdas[i]))
    
fig = plt.figure(figsize=(20,4))

gs1 = gridspec.GridSpec(1, 3)

axes = list()
for i in range(len(outcome_probs)):
    axes.append(fig.add_subplot(gs1[i]))
    axes[i].bar(outcome_counts, outcome_cdf[i])
    axes[i].set_title("Lambda = " + str(lambdas[i]))
    
plt.show()
```

### Fazendo a aproximação

```python
from scipy.stats import binom

outcome_counts = linspace(0,20,21)

outcome_binom = binom.pmf(outcome_counts,20,0.05)
outcome_poisson = poisson.pmf(outcome_counts, 20*0.05)

red_dots = plt.scatter(outcome_counts, outcome_binom, color='r',label='X~binomial(20,0.05)')
blue_dots = plt.scatter(outcome_counts, outcome_poisson, color='b',label='Y~poisson(20*0.05)')

plt.legend(handles=[red_dots, blue_dots])

plt.show()
```
